## Урок 4. Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема

#### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

Найдем среднее значение для равномерного распределения.

$M(X) = \frac{a + b}{2} = \frac{200 + 800}{2} = 500$.

Вычислим дисперсию для равномерного распределения.

$D(X) = \frac{(b - a)^2}{12} = \frac{(800 - 200)^2}{12} = 30000$.

#### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Из условия задачи нам известны: дисперсия $D(X) = 0,2$ и левая граница величины $a = 0,5$.

Из формулы $D(X) = \frac{(b - a)^2}{12}$ вычислим правую границу величины B.

$0,2= \frac{(b - 0,5)^2}{12}$;     $0,2*12= {(b - 0,5)^2}$;      ${2,4}^{0,5} = (b - 0,5)$;      $b = {2,4}^{0,5}+0,5 = 2,04919$

Зная обе границы, вычислим среднее значение.

$M(X) = \frac{a + b}{2} = \frac{0,5 + 2,04919}{2} = 1,2746$.

#### 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-((x+2)**2) / 32)) Найдите: а). M(X) б). D(X) в). std(X) (среднее квадратичное отклонение).

Формула плотности распределения вероятности имеет вид $f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}}$.

Из формулы определим, что $M(X)= -2$; $D(X) = 16$; $std(X)= {D(X)}^{0,5} = {16}^{0,5} = 4$.

#### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
