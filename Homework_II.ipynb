## Урок 2. Дискретные случайные величины. Закон распределения вероятностей. Биномиальный закон распределения. Распределение Пуассона

#### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Поскольку вероятность попадания с первого раза составляет 80%, а также количество испытаний относительно невелико, то для решения подходит формула Бернулли.

$p(X=k) = C_n^kp^kq^{n-k}$

При этом:
    вероятность наступления события- p = 0,8,
    вероятность не попасть в мишень- q = 1 - p = 0,2,
    число наступлений событий- k = 85,
    количество испытаний- n = 100. 

$p(k=85) = C_n^kp^kq^{n-k} = C_{100}^{85}p^{85}q^{100-85} = \frac{100!}{85!(100-85)!} \cdot 0.8^{85} \cdot 0.2^{15} = 0,0481$ ~ 5%.

#### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

Под условия данной задачи подходит формула Пуассона, так как вероятность наступления мала, а также имеется среднее значение на единицу измерения.

$P_m \approx \frac{\lambda^m}{m!}e^{-\lambda}$, где вероятность наступления события- p = 0,0004, $\lambda = p*n = 0,0004*5000 = 2$.

Тогда вероятность того, что ни одна из лампочек не перегорит в первый день- $P_0 \approx \frac{2^0}{0!}e^{-2}=0,13534$ ~ 14%.

Вероятность того, что перегорят две лампочки- $P_2 \approx \frac{2^2}{2!}e^{-2}=0,27067$ ~ 27%.

#### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Для решения задачи подходит формула Бернулли.

Вероятность наступления события- p = 0,5,
вероятность противоположного исхода- q = 1 - p = 0,5,
число наступлений события- k =70,
количество испытаний- n = 144. 

$p(k=70) = C_n^kp^kq^{n-k} = C_{144}^{70}p^{70}q^{144-70} = \frac{144!}{70!(144-70)!} \cdot 0.5^{70} \cdot 0.5^{74} = 0,06281$ ~ 6%.

#### 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Вероятность того, что все четыре мяча будут белыми равна произведению вероятностей извлечения по два белых шара из каждого ящика.

$\frac{7}{10}*\frac{7-1}{10-1}*\frac{9}{11}*\frac{9-1}{11-1}=\frac{3024}{9900}=0,30545$ или ~ 31%.

Извлечь ровно два белых мяча возможно в трех ситуациях: при извлечении двух белых шаров из первого ящика и двух черных из второго, при извлечении двух белых шаров из второго ящика и двух черных из первого или, если из каждого ящика извлекут по одному белому и одному черному шару.

Вероятность извлечения белых мячей только из первого ящика- $\frac{7}{10}*\frac{7-1}{10-1}*\frac{2}{11}*\frac{2-1}{11-1}=\frac{84}{9900}=0,00848$ или ~ 0,8%.

Вероятность извлечения белых мячей только из второго ящика- $\frac{9}{11}*\frac{9-1}{11-1}*\frac{3}{10}*\frac{3-1}{10-1}=\frac{84}{9900}=0,00848$ или ~ 0,8%.

Вероятность того, что хотя бы один из мячей при извлечении будет белым равна разности полной вероятности и произведения вероятностей того, что при извлечении мячи будут черного цвета.

$1-\frac{3}{10}*\frac{3-1}{10-1}*\frac{2}{11}*\frac{2-1}{11-1}=1-\frac{7}{9900}=0,99929$ или ~ 100%.
