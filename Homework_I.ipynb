## Урок 1. Случайные события. Условная вероятность. Формула Байеса. Независимые испытания

##### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

а) В колоде содержится $\frac{52}{4}=13$ карт крестовой масти.

Вероятность извлечь первую карту крестовой масти- $\frac{13}{52}=\frac{1}{4}$;

второй крестовой карты- $\frac{13-1}{52-1}=\frac{12}{51}=\frac{4}{17}$;

третьей карты- $\frac{12-1}{51-1}=\frac{11}{50}$;

четвертой карты крестовой масти- $\frac{11-1}{50-1}=\frac{10}{49}$.

Вероятность извлечения всех четырех карт одной масти будет равна произведению вероятностей последовательного извлечения
первой, второй, третьей и четвертой карт- $\frac{1}{4}*\frac{4}{17}*\frac{11}{50}*\frac{10}{49}=\frac{1320}{499800}=0,00264$ или ~ 0,3%.

б) Всего в колоде содержится 4 туза. 

Вероятность при первом отборе не извлечь туз- $\frac{48}{52}=\frac{12}{13}$;

при втором- $\frac{48-1}{52-1}=\frac{47}{51}$;

при третьем- $\frac{47-1}{51-1}=\frac{46}{50}=\frac{23}{25}$;

при четвертом- $\frac{46-1}{50-1}=\frac{45}{49}$.

Вероятность извлечения хотя бы одного туза при отборе четырех карт будет представлена разностью абсолютной вероятности и произведению вероятностей не извлечь туз при каждом из четырех отборов- $1-\frac{12}{13}*\frac{47}{51}*\frac{23}{25}*\frac{45}{49}=1-0,71874=0,281$ или ~ 28%.

##### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Поскольку код представляет собой одновременное нажатие цифр от 0 до 9, следовательно число возможных вариантов набора будет представлять собой количество сочетаний из трех элементов по 10.

$C_n^k = \frac{n!}{k!(n - k)!}=\frac{10!}{3!(10 - 3)!}=\frac{3628800}{6*5040}=120$.

Вероятность подбора правильного кода с первой попытки будет равна $\frac{1}{120}=0,00833$ или ~ 0,8%.

#### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Вероятность извлечения трех окрашенных деталей (по аналогии с решением в первом задании) равна $\frac{9}{15}*\frac{8}{14}*\frac{7}{13}=\frac{84}{455}=0,18461$ или ~ 18%.

#### 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Принцип решения совпадает с решениями второго и третьего заданий.

$\frac{2}{100}*\frac{1}{99}=0,0002$ или ~ 0,02%.
